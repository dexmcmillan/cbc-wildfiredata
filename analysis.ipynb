```python
import pandas as pd
import glob
import datetime
import pytz
```

```python
csvfiles = glob.glob("data/*.csv")

dfs = []

for file in csvfiles:
    df = pd.read_csv(file, index_col="Unnamed: 0")
    
    dfs.append(df)
    
data = pd.concat(dfs)
```

```python
data.head(3)
```

```python
data["timestamp"] = pd.to_datetime(data["timestamp"])
```

```python
biggest_now = data[data["timestamp"] == data["timestamp"].max()].sort_values("hectares", ascending=False).head(5)["hectares"].to_list()

biggest_now
```

```python
pivot = data.pivot_table(index="timestamp", values="AREA", columns="UID", aggfunc="sum").resample("d").mean()

pivot
```
